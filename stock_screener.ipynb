```python
"""
This code block is used to create the universe of companies that you can look
at. In my article, I describe this process as short listing companies. Some
people will stop here, but I beleive we can go further.
"""

from finvizfinance.screener.valuation import Valuation

# Creating a Valuation object from finviz's API
valuation = Valuation()

"""
The filters (keys) below are available for the .set_filter() method.
Unfortunately there are too many correspoding options (values) for each key.
To get the options like 'Mid ($2bln to $10bln)', go to
https://finviz.com/screener.ashx . From there you will be able to find the
options that are available for each key.
"""

# Possible filter: ['Exchange', 'Index', 'Sector', 'Industry', 'Country',
# 'Market Cap.', 'P/E', 'Forward P/E', 'PEG', 'P/S', 'P/B', 'Price/Cash',
# 'Price/Free Cash Flow', 'EPS growththis year', 'EPS growthnext year',
# 'EPS growthpast 5 years', 'EPS growthnext 5 years', 'Sales growthpast 5 years',
# 'EPS growthqtr over qtr', 'Sales growthqtr over qtr', 'Dividend Yield',
# 'Return on Assets', 'Return on Equity', 'Return on Investment', 'Current Ratio',
# 'Quick Ratio', 'LT Debt/Equity', 'Debt/Equity', 'Gross Margin',
# 'Operating Margin', 'Net Profit Margin', 'Payout Ratio',
# 'InsiderOwnership', 'InsiderTransactions', 'InstitutionalOwnership',
# 'InstitutionalTransactions', 'Float Short', 'Analyst Recom.', 'Option/Short',
# 'Earnings Date', 'Performance', 'Performance 2', 'Volatility', 'RSI (14)',
# 'Gap', '20-Day Simple Moving Average', '50-Day Simple Moving Average',
# '200-Day Simple Moving Average', 'Change', 'Change from Open',
# '20-Day High/Low', '50-Day High/Low', '52-Week High/Low', 'Pattern',
# 'Candlestick', 'Beta', 'Average True Range', 'Average Volume',
# 'Relative Volume', 'Current Volume', 'Price', 'Target Price', 'IPO Date',
# 'Shares Outstanding', 'Float']



# Set filters
filters_dict = {'Market Cap.':'+Mid (over $2bln)',
                'Country':'USA',
                'Debt/Equity': 'Under 0.3',
                'Return on Investment': 'Over +20%'}

valuation.set_filter(filters_dict=filters_dict)
df_valuation = valuation.screener_view()
df_valuation
```

```python
"""
This code block is the actual screening process. From here we disagregate the price
of each company from the short list of companies stored in df_valuation.

Why am I doing this form of screening? I beleive that if we can disagregate
the price of a stock, we can get a better pictrue of the strenghts and
weaknesses of the company.

If the quarterly financial data makes sense, then the stock should be fairly
valued. If not, we will see deviations.
"""
import pandas as pd
from finvizfinance.quote import Statements

# This is the data frame which will store all the relevant data for the companies
df_financials_summary = pd.DataFrame(columns=['Ticker',
                                              'Price/Share',
                                              'Net Income',
                                              'Total Revenue',
                                              'Total Assets',
                                              'Total Equity',
                                              'Net Profit Margin (annualized)',
                                              'Asset Turnonver',
                                              'Leverage',
                                              'Normalized ROE',
                                              'Equity/share',
                                              'EPS',
                                              'Normalized P/E',
                                              'Disagregated Price',
                                              'Valuation',
                                              'P/FV'
                                              ])

# Looping through each ticker
for ticker in df_valuation['Ticker']:
  try:
    stock_statemenets = Statements()
    df_is = stock_statemenets.get_statements(ticker, statement='I', timeframe='Q')
    df_bs = stock_statemenets.get_statements(ticker, statement='B', timeframe='Q')
  except:
    continue

  price = df_valuation.loc[df_valuation['Ticker'] == ticker, 'Price'].values[0]

  # Income Statement values
  net_income, total_revenue, shares_outstanding, price_to_earnings = 0, 0, 0, 0
  # Balance Sheet values
  assets, equity = 0, 0

  # Determines how much far back in time you want to collect data
  is_cols = min(20,len(df_is.columns))
  bs_cols = min(20,len(df_bs.columns))

  # Obtaining each data fields' aggregated total for the specified time period
  for i in range(is_cols):
    try:
      net_income += float(df_is.loc['Net Income'].values[i].replace(',',''))  # Access the value directly
      total_revenue += float(df_is.loc['Total Revenue'].values[i].replace(',',''))  # Access the value directly
      shares_outstanding += float(df_is.loc['Shares Outstanding'].values[i].replace(',',''))  # Access the value directly
      price_to_earnings += float(df_is.loc['Price To Earnings Ratio'].values[i].replace(',',''))  # Access the value directly
    except:
      continue
  for i in range(bs_cols):
    try:
      assets += float(df_bs.loc['Total Assets'].values[i].replace(',',''))  # Access the value directly
      equity += float(df_bs.loc['Total Equity'].values[i].replace(',','')) # Access the value directly
    except:
      continue



  try:
    net_income_avg = net_income/is_cols
    total_revenue_avg = total_revenue/is_cols
    assets_avg = assets/bs_cols
    equity_avg = equity/bs_cols
    shares_outstanding_avg = shares_outstanding/is_cols
  except ZeroDivisionError:
    continue

  # ROE components using DuPont Identity
  try:
    # Annualize net_profit margin
    net_profit_margin = (net_income_avg*4)/total_revenue_avg
    asset_turnover = total_revenue_avg/assets_avg
    leverage = assets_avg/equity_avg
  except ZeroDivisionError:
    continue

  #EPS components
  roe = net_profit_margin * asset_turnover * leverage
  equity_per_share = equity_avg/shares_outstanding_avg
  eps = roe * equity_per_share

  #Disagregated Price
  price_to_earnings_avg = price_to_earnings/is_cols
  disagregated_price = eps * price_to_earnings_avg

  # Create the new row as a DataFrame
  new_row = pd.DataFrame([[ticker,
                         price,
                         int(net_income_avg)/100,
                         int(total_revenue_avg)/100,
                         int(assets_avg)/100,
                         int(equity_avg)/100,
                         net_profit_margin,
                         asset_turnover,
                         leverage,
                         roe,
                         equity_per_share,
                         eps,
                         price_to_earnings_avg,
                         round(disagregated_price, 2),
                         'Overvalued' if (price > disagregated_price + (0.05*disagregated_price)) else 'Undervalued' if (price < disagregated_price - (0.05 * disagregated_price)) else 'Fair Valued',
                         round(price/disagregated_price, 2)]],
                       columns=df_financials_summary.columns)

  # Only concatenate if the new row has valid (non-NA) values
  if not new_row.isnull().all().all():  # Check if the row has all NA values
    df_financials_summary = pd.concat([df_financials_summary, new_row], ignore_index=True)


# Display the final DataFrame
df_financials_summary
df_financials_summary.to_csv('midcap_screen.csv')
```
